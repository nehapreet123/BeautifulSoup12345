# 1-Please visit https://www.cnbc.com/world/?region=world and scrap-

a) headings

b) date

c) News link



```python
import requests
from bs4 import BeautifulSoup
```

```python
#requesting the website
page = 'https://www.cnbc.com/world/?region=world'
response=requests.get(page)
print('The response code is : ',response)

#parse the HTML documents
soup=BeautifulSoup(response.content, 'html.parser')


#extracting the news heading
headings=soup.find_all('h3')

#Display the headings
for heading in headings:
    print(heading.text,"\n")
```
